```python
import math
```

```python
def Combination(k,n): 
  '''Функция для сочетний
  k - количество элементов в наборе
  n - общее число элементов
  '''
  return int(math.factorial(n) / (math.factorial(k) * math.factorial(n - k)))
```

```python
def Razmeshenie(k,n): 
  '''Функция для размещения
  k - количество элементов в наборе
  n - общее число элементов
  '''
  return int(math.factorial(n) / math.factorial(n - k))
```

```python
def Bernully(k,n,p):
  '''P(X=k) = C_n ^k    *〖p 〗^k*q^(n-k)
позволяющая находить вероятность появления события 
A определённое количество раз при любом числе независимых испытаний
	k - число наступления события (дискретная величина из отрезка [0, n])
	n -  число испытаний 
	p - вероятность наступления события A в  независимых испытаниях, 
	q = 1 – p. Противоположная вероятность

  '''
  return Combination(k,n) * p ** k * (1 - p) ** (n - k)
```

```python
def Puasson(k,n,p):
  '''
P(k)≈λ^k/k!*e^(-λ)

k — количество событий,
λ  — математическое ожидание случайной величины (среднее количество событий за фиксированный промежуток времени),
k! обозначает факториал числа 
e=2,718.. — основание натурального логарифма
вероятность, что событие наступит m раз из n

λ. Это средняя интенсивность наступления события на некоторую единицу измерения
λ = p*k 

Распределе́ние Пуассо́на — распределение дискретного типа случайной величины, представляющей собой число событий, 
произошедших за фиксированное время, при условии, что данные события происходят с некоторой фиксированной средней 
интенсивностью и независимо друг от друга.

  '''
  λ = p*n
  return (λ) ** k / math.factorial(k) * math.exp(-λ)
```

1) Вероятность того, что стрелок попадет в мишень, выстрелив один раз, равна 0.8. Стрелок выстрелил 100 раз. Найдите вероятность того, что стрелок попадет в цель ровно 85 раз.

```python
import numpy as np
a = np.ones(800)
a
```

```python
b =np.zeros(200)
b
```

```python
full = np.hstack([a, b])
full
```

```python
np.random.shuffle(full)
full
```

```python
trialsData = full.copy()
trialsData
```

```python
trialsQty = 9999
```

```python
for i in range (trialsQty):
  np.random.shuffle(full)
  trialsData = np.vstack([trialsData, full])
```

```python
np.shape(trialsData)
```

```python

resultsTrials = np.sum(trialsData[:,:100], axis = 1)
np.shape(resultsTrials)
```

```python
np.shape(resultsTrials)
```

```python
resultsTrials85 = resultsTrials[resultsTrials == 85]
resultsTrials85
```
